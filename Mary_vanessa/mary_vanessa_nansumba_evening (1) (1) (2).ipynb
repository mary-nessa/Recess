```python
import pandas as pd
dataset = {
    'days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'dates': ['02-06-2022', '03-06-2022', '04-06-2022', '05-06-2022', '06-06-2022', '07-06-2022', '08-06-2022']
}
s = pd.Series(dataset)
days = []
index = -1
for j in s['days']:
    index = index + 1
    for i in s['dates']:
        days.append(i)
    print(j,":", days[index])
```

```python
#EXERCISE
#DATA ANALYSIS WITH PANDAS
#SERIES
#N1.Write a Pandas program to compare the elements of two Pandas series. 
#Sample Series [4,65,436,3,9], [7,0,3,897,9]

#WE IMPORT PANDAS INORDER TO USE IT
import pandas as pd
#s is for series.....
#  Series
s1 = pd.Series([4, 65, 436, 3, 9])
s2 = pd.Series([7, 0, 3, 897, 9])
# Compare the elements of the two Series using 
#the `==` operator which compare each element of `s1` with the corresponding element in `s2`
comparison = s1 == s2

# Print the comparison result this returns either true or false
print(comparison)
```

```python
#Number two
#Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
#Sample Series: [2,4,6,8,14], [1,3,5,7,9]
import pandas as pd

# Series to use
s1 = pd.Series([2, 4, 6, 8, 14])
s2 = pd.Series([1, 3, 5, 7, 9])

# Addition of series
addition = s1.add(s2)

# Results
print("Addition Result:")
print(addition)

# Subtraction of series
subtraction = s1.subtract(s2)

print("\nSubtraction Result:")
print(subtraction)

# Multiplication
multiplication = s1.multiply(s2)

print("\nMultiplication Result:")
print(multiplication)

# Division
division = s1.divide(s2)

print("\nDivision Result:")
print(division)
```

```python
#Number 5
#Write a Pandas program to display the most frequent value in the given series 
#and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)
import pandas as pd

# Read the dataset (assuming it is a CSV file)
Africa_df = pd.read_csv('/Users/mojca/Desktop/Mary-Nansumba/AfricaCupofNationsMatches.csv')

# Extract the 'HomeTeamGoals' column as a series
series = Africa_df['HomeTeamGoals']

# Find the most frequent value
most_frequent = series.mode()[0]

# Replace everything else with 'Other'
series = series.apply(lambda x: "Other" if x != most_frequent else x)

# Print the modified series
print(series)
```

```python
# Number 18
#Write a Pandas program to iterate over rows in your DataFrame.
import pandas as pd

# Reading the CSV file
afri_cup = pd.read_csv('/Users/mojca/Desktop/Mary-Nansumba/AfricaCupofNationsMatches.csv')

# Creating a set to store unique rows
unique_rows = set()

# Iterating over rows in the DataFrame
for index, row in afri_cup.iterrows():
    # Convert the row to a tuple
    row_tuple = tuple(row)

    # Checking if the row is already in the set
    if row_tuple in unique_rows:
        continue

    # Adding the row to the set
    unique_rows.add(row_tuple)
    
    # Printing the unique row
    print("Row index:", index)
    print("Row data:")
    print(row)
    print("------------------------------")
```
