```python
off = ['YEAR', 'DISTRICT ', 'SCHOOL', 'TOTAL CANDIDATES']
for col in df1.columns:
    if col not in off:
        if np.issubdtype(df1[col].dtype, np.number):
            df1[col].plot(kind='pie')
            plt.title(col)
            plt.show()
```

```python
import numpy as np
# Identify numeric columns with NaN values
numeric_columns_with_nan = []
for column in df1.columns:
    if np.issubdtype(df1[column].dtype, np.number) and df1[column].isnull().values.any():
        numeric_columns_with_nan.append(column)

# Calculate median for each numeric column and fill NaN values
for column in numeric_columns_with_nan:
    column_median = df1[column].median()
    df1[column].fillna(column_median, inplace=True)

# Verify if NaN values have been filled
nan_values_remaining = df1.isnull().sum().sum()
print("Number of NaN values remaining:", nan_values_remaining)

```
